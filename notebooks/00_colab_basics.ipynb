### 1. Colab Runtime Information

Every Colab session runs on a fresh virtual machine. Let's check
the Python version and confirm we are running inside Colab.

```python
#@title Check Colab runtime info
import sys
import os

print(f"Python version: {sys.version}")
if "google.colab" in sys.modules:  # running on Colab
    print("Running on Google Colab.")
else:
    print("Not running on Google Colab.")
print(f"Current directory: {os.getcwd()}")
```

### 4. Connecting to Google Drive

Google Drive provides persistent storage that survives across
Colab sessions. Mounting Drive makes it available at
`/content/drive`. You will be prompted to authorize access.

```python
#@title Mount Google Drive
from google.colab import drive

drive.mount("/content/drive")
```

### 5. Saving and Verifying an Artifact

Once Drive is mounted, you can read and write files as if they
were on a local disk. The following cells create a small JSON
artifact, save it to Drive, and read it back to verify.

```python
#@title Create and save a small artifact
import json
from pathlib import Path

artifact = {
    "source": "00_colab_basics",
    "message": "Hello from Google Colab!",
    "packages": ["httpx"],
}
drive_dir = Path("/content/drive/MyDrive/colab_artifacts")
drive_dir.mkdir(parents=True, exist_ok=True)
artifact_path = drive_dir / "artifact.json"
with open(artifact_path, "w") as f:
    json.dump(artifact, f, indent=2)
print(f"Artifact saved to: {artifact_path}")
```

```python
#@title Read back and verify the artifact
with open(artifact_path) as f:
    loaded = json.load(f)
print(json.dumps(loaded, indent=2))
print("Artifact verified successfully!")
```
